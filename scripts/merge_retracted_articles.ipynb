```python
from glob import glob
import pandas as pd
import os
```

### merge the resulting tab delimited retracted article files into one large retracted_articles.csv file

```python
adf = None
for fn in glob('retracted_articles/*.txt'):
    #print(fn)
    if fn.find('cites') != -1:
        continue
    df = pd.read_csv(fn, delimiter='\t')
    ref_id = int(os.path.basename(fn).split('.')[0])
    df['ref_id'] = ref_id
    if adf is None:
        adf = df
    else:
        adf = pd.concat([adf, df])
```

```python
adf.head()
```

```python
del adf['Unnamed: 57']
adf.sort_values(['ref_id'], inplace=True)
adf
```

```python
adf.to_csv('retracted_articles.csv.gz', index=False, compression='gzip')
```

### merge the resulting tab delimited retracted article citation files into one large citations.csv file

```python
adf = None
for fn in glob('retracted_articles/*.txt'):
    #print(fn)
    if fn.find('cites') == -1:
        continue
    try:
        df = pd.read_csv(fn, delimiter='\t')
    except Exception as e:
        print(e)
        df = pd.read_csv(fn, delimiter='\t', error_bad_lines=False, engine='python')
    ref_id = int(os.path.basename(fn).split('_')[0])
    df['ref_id'] = ref_id
    if adf is None:
        adf = df
    else:
        adf = pd.concat([adf, df])
```

```python
adf.head()
```

```python
del adf['Unnamed: 57']
adf.sort_values(['ref_id'], inplace=True)
adf
```

```python
adf.to_csv('citations.csv.gz', index=False, compression='gzip')
```
